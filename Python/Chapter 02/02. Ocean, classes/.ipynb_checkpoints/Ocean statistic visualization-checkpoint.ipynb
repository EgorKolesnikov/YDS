```python
%matplotlib inline
import numpy as np
import collections
import matplotlib.pyplot as plt
import warnings
import pandas
warnings.filterwarnings('ignore')

def work(number):
    predators = []
    victims = []
    with open('result' + number + '.txt', 'r') as file:
        for line in file:
            t = line.split()
            predators.append(t[0])
            victims.append(t[1])

        ticks = np.linspace(0, len(predators), len(predators))
        statistic = [float(p) / float(v) for p, v in zip(predators, victims)]

        fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 4))
        axes[0].plot(ticks, predators, 'ro')
        axes[0].plot(ticks, predators, 'r', label='Predators')
        axes[0].plot(ticks, victims, 'bo')
        axes[0].plot(ticks, victims, 'b', label='Victims')
        axes[0].set_xlabel("Time state")
        axes[0].set_ylabel("predators, victims")
        axes[0].set_title("Ocean life statistic")
        axes[0].legend(loc='upper_left')
        axes[1].plot(ticks, statistic, 'go', ticks, statistic, 'g')
        axes[1].set_xlabel("Time state")
        axes[1].set_ylabel("Ration predators to victims")
        axes[1].set_title("Ocean life statistic")

        

for i in xrange(1, 13, 1):
    work(str(i))
```
